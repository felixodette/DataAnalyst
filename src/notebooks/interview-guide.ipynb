```python
import pandas as pd 
import plotly.express as px
import klib as klib
```

```python
# read in data
df = pd.read_csv('interview.csv')

# prepare bio data information
bio_data = ['Gender', 'Location Country', 'City', 'At which career level are you?', 'Which membership level would you be interested in?']
user_bio = df[bio_data]
user_bio = user_bio.rename(columns={'Location Country': 'Country', 'At which career level are you?':'Career Level', 'Which membership level would you be interested in?':'MemType Interest'})

# group bio data by gender
gender = pd.DataFrame(user_bio.groupby('Career Level')['Gender'].value_counts()).rename(columns={'Gender': 'Number'}).reset_index()
```

# Bio Data Distribution


```python
# plot sunurst of biodata
fig=px.sunburst(gender, path=['Career Level','Gender'], values='Number', title='Fig 1: Bio Data Distribution')
fig.show()
```

# Correlation
Definition: Correlation is a statistical measure that expresses the extent to which two variables are linearly related (meaning they change together at a constant rate). It’s a common tool for describing simple relationships without making a statement about cause and effect.

Highly correlated features include:
* 0.94 Past successes enhances or deters your ability to lead innovation - and - I make sure my time is orgnanized for maximized benefit
* 0.91 I make sure my time is organized for the maximum benefit - and - Once I make a decision I’m pleased with my choice regardless of the result

Lowly correlated featuers include:
* -0.94 You use disruptive innovation to make meaningful impact at work and in society. - and - Join as a member
* -0.88 You measure the input, workflow and output of innovation in your organisation and industry. - A lot of what I do feels trivial

```python
klib.corr_plot(df)
```
